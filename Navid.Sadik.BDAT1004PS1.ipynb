```python
// Question 3

using System;
using System.Runtime.ExceptionServices;

public class Question3
{
    static double TriangleArea(double a, double b, double c)
    {

        double s = (a + b + c) / 2;
        double area=  Math.Sqrt(s * (s - a) * (s - b) * (s - c));
        return area;
    }
    public static void Main(String[]args)
    {
        Console.Write("Enter the length of first side ===> ");
        double a = double.Parse(Console.ReadLine());
        Console.Write("Enter the length of second side ===> ");
        double b = double.Parse(Console.ReadLine());
        Console.Write("Enter the length of third side ===> ");
        double c = double.Parse(Console.ReadLine());
        Console.WriteLine("Area of the triangle is: " + TriangleArea(a,b,c));

    }
}
```

```python
//Question 4

using System;
using System.Runtime.ExceptionServices;

public class Question4
{
    public static void Main()
    {
        int[] array1 = new int[5];
        int[] array2 = new int[5];
        int[] array3 = new int[5];
        int i, j = 0, k = 0, n;

        Console.Write("Input the number of elements for the array :");
        n = Convert.ToInt32(Console.ReadLine());

        Console.Write("Input {0} elements in the array :\n", n);
        for (i = 0; i < n; i++)
        {
            Console.Write("element - {0} : ", i);
            array1[i] = Convert.ToInt32(Console.ReadLine());
        }

        for (i = 0; i < n; i++)
        {
            if (array1[i] % 2 == 0)
            {
                array2[j] = array1[i];
                j++;
            }
            else
            {
                array3[k] = array1[i];
                k++;
            }
        }

        Console.Write("\nThe Even elements are : \n");
        for (i = 0; i < j; i++)
        {
            Console.Write("{0} ", array2[i]);
        }

        Console.Write("\nThe Odd elements are :\n");
        for (i = 0; i < k; i++)
        {
            Console.Write("{0} ", array3[i]);
        }
        Console.Write("\n\n");
    }
}
```

```python
//Question 5

using System;
public class Question5
{
    static bool Inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        return x >= x1 && x <= x2 && y >= y1 && y <= y2;
    }

    public static void Main(string[] args)
    {
        Console.WriteLine("Present the value of x1: ");
        double x1 = double.Parse(Console.ReadLine());
        Console.WriteLine("Present the value of y1: ");
        double y1 = double.Parse(Console.ReadLine());
        Console.WriteLine("Present the value of x2: ");
        double x2 = double.Parse(Console.ReadLine());
        Console.WriteLine("Present the value of y2: ");
        double y2 = double.Parse(Console.ReadLine());
        Console.WriteLine($"The rectangle coordinates are: ({x1},{y1}),({x2},{y2})");

        Console.WriteLine("Present x axis value of the point: ");
        double x = double.Parse(Console.ReadLine());
        Console.WriteLine("Present y axis value of the point: ");
        double y = double.Parse(Console.ReadLine());

        if (Inside(x, y, x1, y1, x2, y2))
        {
            Console.WriteLine("The point is inside the rectangle");
        }
        else
        {
            Console.WriteLine("The point is outside of the rectangle");
        }
    }
}
```

```python
//Question 6

userInput = input("Enter word :\n")

vowels = ['a','e','i','o','u']
def pig(userInput): 
    first = userInput[0]
    
    if first in vowels: 
         userInput = userInput.lower()
         userInput += "way" 
         return userInput
    else: 
        userInput = userInput.lower()

        for letter in userInput:
            if letter in vowels:
                index_value = userInput.index(letter)
                break

        userInput = userInput[index_value:] +userInput[:index_value]+ "ay" 
        return userInput 

print(pig(userInput))
```

```python
//Question8

with open('currencies.txt', 'rt') as f:
    lines = f.readlines()

currencyDict = {}
for line in lines:
    parsed = line.split("\t")
    currencyDict[parsed[0]] = parsed[1]

amount = int(input("Enter amount: \n "))
print("Enter your currency: (Available options:\n)")
[print(item) for item in currencyDict.keys()]
currency = input("Please enter one of the options:")
print(
    f"{amount} is equal to {amount *float (currencyDict[currency])} USD")
```
